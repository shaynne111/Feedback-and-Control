```python
import control as ct
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define Transfer Function G(s) = (s + 1) / (s^4 + 4s^3 + 7s^2 + 6s + 3)
num = [1, 1]
den = [1, 4, 7, 6, 3]
G = ct.TransferFunction(num, den)
```

```python
# Performance Metrics
info = ct.step_info(G)
poles = ct.poles(G)
zeros = ct.zeros(G)
```

```python
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
print(f"Rise Time: {info['RiseTime']:.4f} s")
print(f"Settling Time (2%): {info['SettlingTime']:.4f} s")
print(f"Overshoot: {info['Overshoot']:.4f} %")
```

```python
# Visualization
plt.figure(figsize=(12, 10))
```

```python
# Step Response
plt.subplot(2, 2, 1)
t, y = ct.step_response(G)
plt.plot(t, y)
plt.title('Step Response')
plt.grid(True)
```

```python
# Bode Plot
plt.subplot(2, 2, 2)
ct.bode_plot(G)
plt.subplot(2, 2, 3) # Note: bode_plot handles subplots internally in some versions
```

```python
# Pole-Zero Map
plt.subplot(2, 2, 4)
ct.pzmap(G)
plt.show()
```
