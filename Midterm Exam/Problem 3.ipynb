```python
import control
import matplotlib.pyplot as plt
import numpy as np
```

```python
# G2(s) = (2s^2 + s + 3) / (8s^4 + 8s^3 + 24s^2 + 12s)
num = [2, 1, 3]
den = [8, 8, 24, 12, 0]
```

```python
# Create system object
sys = control.TransferFunction(num, den)
```

```python
# Stability & Pole-Zero Analysis
poles = control.poles(sys)
zeros = control.zeros(sys)
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
```

```python
# Metric Extraction
try:
    info = control.step_info(sys)
    for key, value in info.items():
        print(f"{key}: {value:.4f}")
except Exception as e:
    print("Metric extraction failed (system may be unstable or marginal)")
```

```python
# Visualization Suite
plt.figure(figsize=(12, 10))
```

```python
# 1. Step Response
plt.subplot(2, 2, 1)
t, y = control.step_response(sys)
plt.plot(t, y, color='#007B8C', lw=2)
plt.title('Step Response')
plt.grid(True, alpha=0.3)
```

```python
# 2. Pole-Zero Map
plt.subplot(2, 2, 2)
control.pzmap(sys)
plt.title('Pole-Zero Map')
```

```python
# 3. Bode Magnitude
plt.subplot(2, 2, 3)
mag, phase, omega = control.bode(sys, plot=False)
plt.semilogx(omega, 20 * np.log10(mag), color='#20c997')
plt.title('Bode Magnitude')
plt.ylabel('Magnitude (dB)')
plt.grid(True, which='both', alpha=0.3)
```

```python
# 4. Bode Phase
plt.subplot(2, 2, 4)
plt.semilogx(omega, np.rad2deg(phase), color='#f59e0b')
plt.title('Bode Phase')
plt.ylabel('Phase (deg)')
plt.grid(True, which='both', alpha=0.3)
```

```python
plt.tight_layout()
plt.show()
```
